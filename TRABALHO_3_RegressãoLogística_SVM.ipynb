Ela é usada para realizar ajustes de hiperparâmetros durante o treinamento de um modelo. O grid search automatiza o processo de encontrar hiperparâmetros ideais, economizando esforço humano em comparação com o ajuste manual, mas pode até ser mais custoso do ponto de vista de desempenho, pois testa todas as combinações possíveis e retorna a que obteve melhor desempenho.
Agora, aplique o GridSearch do Scikit-Learn (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) para ajudar a identificar os melhores hiperparâmetros para o seu modelo. 1. Aplique o algoritmo SVM na base de dados ‘credit.pkl’.

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pickle

with open('credit.pkl', 'rb') as f:
    X_credit_train, y_credit_train, X_credit_test, y_credit_test = pickle.load(f)

print(X_credit_train.shape, X_credit_test.shape)
```

2. Inicialmente treine o SVM com kernel linear, valor do parâmetro C = 1.0 e ‘random_state =1’

```python
svm_linear = SVC(kernel='linear', C=1.0, random_state=1)
svm_linear.fit(X_credit_train, y_credit_train)
```

3. Utilize o comando do sklearn accuray_score para calcular a acurácia do seu algoritmo. O resultado deve ser 0.946

```python
y_pred_linear = svm_linear.predict(X_credit_test)
acc_linear = accuracy_score(y_credit_test, y_pred_linear)
print(f"Acurácia (kernel linear): {acc_linear:.3f}") 
```

4. Teste os demais kernels e anote os resultados. Qual o melhor kernel para a sua base de dados?
    * Polinomial
    * Sigmoide
    * rbf

```python

kernels = ['poly', 'sigmoid', 'rbf']

resultados = {}
melhor_acc = 0
melhor_kernel = None

for k in kernels:
    modelo = SVC(kernel=k, C=1.0, random_state=1)
    modelo.fit(X_credit_train, y_credit_train)
    acc = accuracy_score(y_credit_test, modelo.predict(X_credit_test))
    resultados[k] = acc
    print(f"Acurácia (kernel {k}): {acc:.3f}")

    # if simples para armazenar o melhor kernel até o momento
    if acc > melhor_acc:
        melhor_acc = acc
        melhor_kernel = k

print(f"\nMelhor kernel: {melhor_kernel} com acurácia de {melhor_acc:.3f}")
```

5. Aumente o valor do parâmetro C aplicado ao melhor kernel e verifique se há mudanças no resultado do seu SVM.

```python

svm_melhor_c10 = SVC(kernel=melhor_kernel, C=10.0, random_state=1)
svm_melhor_c10.fit(X_credit_train, y_credit_train)
acc_c10 = accuracy_score(y_credit_test, svm_melhor_c10.predict(X_credit_test))
print(f"Acurácia ({melhor_kernel} com C=10): {acc_c10:.3f}")
```

6. O Grid Search (pesquisa em grade) é uma técnica utilizada para melhorar a precisão e a generalização dos modelos de aprendizado de máquina. Ela é usada para realizar ajustes de hiperparâmetros durante o treinamento de um modelo. O grid search automatiza o processo de encontrar hiperparâmetros ideais, economizando esforço humano em comparação com o ajuste manual, mas pode até ser mais custoso do ponto de vista de desempenho, pois testa todas as combinações possíveis e retorna a que obteve melhor desempenho.
Agora, aplique o GridSearch do Scikit-Learn (https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) para ajudar a identificar os melhores hiperparâmetros para o seu modelo. Qual foi a melhor combinação de hiperparâmetros encontrada? O modelo com melhor desempenho foi obtido com os parâmetros ajustados manualmente ou com o GridSearch?

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "C": [0.1, 1, 10],
    "kernel": ["linear", "poly", "rbf", "sigmoid"],
    "gamma": ["scale", "auto"]
}
grid = GridSearchCV(SVC(random_state=1), param_grid, cv=5, scoring="accuracy")
grid.fit(X_credit_train, y_credit_train)

print("\nMelhores parâmetros encontrados pelo GridSearch:")
print(grid.best_params_)

melhor_modelo = grid.best_estimator_
y_pred = melhor_modelo.predict(X_credit_test) 
acuracia = accuracy_score(y_credit_test, y_pred)  
print(f"Acurácia do melhor modelo ajustado pelo GridSearch: {acuracia:.3f}")
```

O GridSearchCV encontrou valores melhores para (C=10), resultando em uma acurácia superior.
Porém, a diferença é pequena, ou seja o modelo já estava muito próximo do ideal.
